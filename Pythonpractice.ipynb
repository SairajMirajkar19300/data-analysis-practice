```python
#Mini project Create a student score calculator where the user enters 5 subjects, and you return total, average, pass/fail, and grade.
def student(score):
  a = input("ENTER 1ST SUB NUMBER:")
  b = input("ENTER 2ND SUB NUMBER:")
  c = input("ENTER 3RD SUB NUMBER:")
  d = input("ENTER 4TH SUB NUMBER:")
  e = input("ENTER 5TH SUB NUMBER:")

  total = int(a) + int(b) + int(c) + int(d) + int(e)
  average = total/5
  print("Total is :", total)
  print("Average is :", average)
  if average >= 90:
    return "Grade - A (PASS)"
  elif average >= 80:
    return "Grade - B (PASS)"
  elif average >= 70:
    return "Grade - C (PASS)"
  elif average >= 60:
    return "D"
  else:
    return "FAIL"

print(student(5))
```
